```python
# Dependencies and Setup
# SQL Alchemy
from sqlalchemy import create_engine

# Pandas
import pandas as pd

# Matplotlib
import matplotlib.pyplot as plt

# NumPy
import numpy as np

import sqlite3
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:postgres@localhost:5432/employeesql_db')
conn = engine.connect()
```

```python
#Create a histogram to visualize the most common salary ranges for employees.
#conn = sqlite3.connect("data/portal_mammals.sqlite")
salaries_df = pd.read_sql("SELECT * FROM salaries", engine)
salaries_df.head()
```

```python
#Data from Titles
titles_df= pd.read_sql("SELECT * FROM title", engine)
titles_df
```

```python
#Data from employees
employees_df= pd.read_sql("SELECT * FROM employees", engine)
employees_df.head()
```

```python
#combine salaries to employees using emp_no
combined1_df= pd.merge(salaries_df, employees_df, on="emp_no", how="inner")
combined1_df.head()                      
```

```python
#combine data from title_id
combined2_df= pd.merge(combined1_df, titles_df, on="emp_title_id", how="inner")
combined2_df.head()   
```

```python
#pull data just for title and salary
grouped_df = combined2_df.groupby("title").mean()
grouped_df.round(2)
```

```python
combined3_df = grouped_df.drop(columns = "emp_no")
combined3_df.round(2)
```

```python
reset_df = combined3_df.reset_index()
reset_df.round(2)
```

```python
#Creat a bar chart
salary_avg = reset_df[['title', 'salary']] 
salary_avg.plot(kind="bar", x='title', y='salary')
plt.xticks(rotation=45)
plt.title("Average Salary by Title")
plt.xlabel("Title")
plt.ylabel("Salary")
plt.show()
plt.tight_layout()
```
